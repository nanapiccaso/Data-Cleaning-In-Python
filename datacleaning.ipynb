```python
#unamed column has percentage of missing values > 50% so we drop that column

df=df.drop(['Unnamed: 8'],axis=1)
```

```python

#checking columns with null values
df[df.isnull().any(axis=1)]

```

```python
#checking for number of duplicates

df.duplicated().sum()
```

```python
#dentifying garbage value

for i in df.select_dtypes(include="object").columns:
    print(df[i].unique())
    print("****"*10)
```

```python
#removing garbarge values from duration column
df['Duration']=df['Duration'].replace([' ','Nan','Inf','-','Not Applicable'], np.nan)

#removing garbage values from country column
df['Country']=df['Country'].str.strip('1')  
df['Country'] = df['Country'].str.replace('US.','USA')
df['Country'] = df['Country'].str.replace('USA','US')

df['Country'] = df['Country'].str.replace('West Germany','Germany')


    
```

```python
print(df.columns.tolist())

#converting votes column to string
df[' Votes '] =  df[' Votes '].astype(str)

#removing garbage values from Votes column
df[' Votes ']=df[' Votes '].str.replace('.','')

#converting votes column to float
df[' Votes ']= df[' Votes '].astype(float)

#converting vote column values back to correct decimal values

def kwame(x):
    if len(str(x)) >=8:
       return x/1000
    else:
        return x/100

df[' Votes ']=df[' Votes '].apply(kwame)


```

```python

#removing garbage values from Income column
df['Income']= df['Income'].str.replace("o","0")

#converting income column to string
df['Income']=df['Income'].astype(str)

#adding commas to currency
df['Income']=df['Income'].str.replace(',','')
df['Income']=df['Income'].str.replace('$','')
df['Income']=df['Income'].str.strip("")
from numpy import int64
df['Income']=df['Income'].replace('nan','0').astype(int64)
df['Income']=df.apply(lambda x: "${:,}".format(x['Income']), axis=1)



```

```python
#removing garbage values from Score column
df['Score']= df['Score'].astype(str)

df['Score']= df['Score'].str.strip('f0++e-0.,')
df['Score']=df['Score'].str.replace("..",".")
df['Score']=df['Score'].str.replace(":",".")
df['Score']=df['Score'].str.replace(",",".")
df['Score']=df['Score'].astype('float')

```

```python
#cleaning Realease year column

df['Release year']=df['Release year'].str.replace("The 6th of marzo, year 1951","6/03/1951")

df['Release year']=df['Release year'].str.replace('34-02-1984',"nan")


#converting to datetime format
df['Release year']=pd.to_datetime(df['Release year'],format="mixed",errors='coerce')

df['Release year']=df['Release year'].dt.strftime('%d-%m-%Y')

```

```python
#changing column names
print(df.columns.tolist())

#Converting column names to string some are not in the right format
df.columns = df.columns.astype(str)

#renaming the columns

df.rename(columns={'IMBD title ID':'ID', 'False':'TITLE','Release year':'RELEASE YEAR','False.1':'GENRE','Duration':'DURATION','Country':'COUNTRY','Content Rating':'CONTENT RATING','Director':'DIRECTOR',
                  'Income':'INCOME',' Votes ':'VOTES','Score':'SCORE'}, inplace= True)

df.head()
```

```python


#cleaning duration column

df.info()
df['DURATION']= df['DURATION'].astype(str)

df['DURATION']=df['DURATION'].str.strip("c")
df['DURATION']=df['DURATION'].str.replace("nan","0")
```

```python
df.isnull().sum()
```

```python
#dropping null values
df.dropna(inplace=True)
df[df.isnull().any(axis=1)]
```

```python
df.info()
```

```python
#change duration to datatype int64

df['DURATION']=df['DURATION'].astype(int64)
```

```python
df.info()
```

```python
df.head(50)
```

```python
#filling cells where duration is 0 with the avegrage of the duration 
df['DURATION'].mean()

def replace_zero(x):
        if x==0:
            return x+(df['DURATION'].mean())
        else:
            return x



df['DURATION']=df['DURATION'].apply(replace_zero)        
        
```

```python
df.head(20)
```

```python
#rounding duration and votes column to 2 decimal places
df['DURATION']=df['DURATION'].apply(lambda x:round(x))
df['VOTES']=df['VOTES'].apply(lambda x:round(x,2))
```
